```python
# Importando as bibliotecas
from textblob import TextBlob
from unidecode import unidecode
```

```python
from googletrans import Translator
```

```python
translator = Translator()
```

```python
#Contadores de sentimentos positivos e negativos - Bolsonaro
numPos = 0
numNeg = 0
total = 0
```

```python
#Aplicando unidecode aos textos do Tweet - Bolsonaro
textPT = df_bolsonaro['Texto'].apply(unidecode)
display(textPT)
```

```python
sentimento_bolsonaro = []

#Traduzindo para o InglÃªs - Bolsonaro
for x in textPT:
    textEN = translator.translate(x, dest="en") #traduzindo
    sentimento = TextBlob(textEN.text) #obtendo o sentimento
    sentimento_bolsonaro.append(sentimento.polarity) #adicionando a uma lista
display(sentimento_bolsonaro)
```

```python
total = len(sentimento_bolsonaro)
display(total)
```

```python
for valor in sentimento_bolsonaro:
    if valor > 0: 
        numPos += 1 
    elif valor < 0: 
        numNeg += 1
```

```python
mediaPos = numPos/total
mediaNeg = numNeg/total
print('Porcentagem de comentÃ¡rios positivos - Bolsonaro: '+str(mediaPos))
print('Porcentagem de comentÃ¡rios negativos - Bolsonaro: '+str(mediaNeg))
```
